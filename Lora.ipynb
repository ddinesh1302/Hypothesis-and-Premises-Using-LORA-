```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r"D:\AI Project\car_dealership.csv")
df.head()
```

```python
intent_registery = {
    "Vehicle Recommendation":
        "This question is about suggesting suitable cars based on budget, usage, and preferences.",

    "Model Comparison":
        "This question is about comparing different car models based on features, pricing, and performance.",

    "Test Drive Scheduling":
        "This question is about assisting customers in booking and managing car test drives.",

    "Financing Advisory":
        "This question is about providing car loan options, EMI details, and financing guidance.",

    "Insurance Guidance":
        "This question is about advising suitable car insurance plans and coverage options.",

    "Trade-In Evaluation":
        "This question is about estimating the value of old vehicles for exchange offers.",

    "Inventory Availability":
        "This question is about checking stock availability of car models and variants.",

    "Variant Selection Advice":
        "This question is about helping customers choose the right variant and trim level.",

    "Fuel Type Advisory":
        "This question is about recommending petrol, diesel, hybrid, or electric vehicles.",

    "After-Sales Support Information":
        "This question is about providing information on service centers, warranties, and post-purchase support."
}

```

```python
hypo_2_intent = {
    v : k for k, v in intent_registery.items()
}
```

```python
df["HYPOTHESIS"] = (
    df["HYPOTHESIS"]
    .str.strip()
    .str.replace('"', '', regex=False)
)
```

```python
df["PREMISE"] = (
    df["PREMISE"]
    .str.strip()
    .str.replace('"', '', regex=False)
)
```

```python
df["intent"] = df["HYPOTHESIS"].map(hypo_2_intent)
df.head()
```

```python
label_encoder = LabelEncoder()
df["label"] = label_encoder.fit_transform(df["intent"])
```
